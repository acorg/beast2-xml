# Phase 2i: Running IQtree Correction

IQ tree is offended by the characters: '/'. The code in this module corrects this.

<details>
    <summary>Click To See A Decription of Parameters</summary>
        <pre>
            <code>
fasta_path: str
    Path to fasta file containing sequences used to generate IQ tree at tree_path.

tree_path: str
    Path to IQ tree file.

  </code>
</pre>

```python
fasta_path = None
tree_path = None
```

```python
# Parameters
fasta_path = "example_data/COVID-19_JN.1/VirusSeq_JN1_alignment.fasta"
tree_path = "Local-Test-BDSKY-Rt_change_times/2025-11-10_13-28-02/initial_trees/iqtree.nwk"
```

```python
from Bio import SeqIO
```

```python
correction_dict = {seq_record.id.replace('/', '_'): seq_record.id for seq_record in SeqIO.parse(fasta_path, 'fasta')}
fh = open(tree_path)
tree = fh.read()
for changed, original in correction_dict.items():
    tree = tree.replace(changed, original)

oh = open(tree_path, 'w')
oh.write(tree)
oh.close()
```
